```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Algerian_forest_fires_cleaned_dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['month' , 'day' , 'year'] , axis=1 , inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## encoding

df['Classes'] = np.where(df['Classes'].str.contains("not fire"), 0, 1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## independent and dependent features

X = df.drop('FWI' , axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
## train test split

from sklearn.model_selection import train_test_split

X_train,X_test , y_train , y_test = train_test_split(X , y , random_state=42 , test_size=0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
## Feature selection

X_train.corr()
```

```python
## Check multicolliniarity

plt.figure(figsize=(12 , 10))
corr = X_train.corr()
sns.heatmap(corr , annot=True)
plt.show()
```

```python
def correlation(dataset , threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i , j]) > threshold:
                col_name = corr_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr
```

```python
corr_features =correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85

X_train.drop(corr_features , axis=1 , inplace=True)
X_test.drop(corr_features , axis=1 , inplace=True)
X_train.shape , X_test.shape
```

```python
## Feature Scaling lets do Standardization

from sklearn.preprocessing import StandardScaler

scaler  = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_train_scaled
```

```python
## Boxplots to understand effect of Standard Scaler

plt.subplots(figsize = (15 , 5))
plt.subplot(1 , 2 , 1)
sns.boxplot(data= X_train)
plt.title("X_train Before Scaling")
plt.subplot(1 , 2, 2)
sns.boxplot(data = X_train_scaled)
plt.title("X_train After Scaling")
plt.show()
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
model = LinearRegression()
model.fit(X_train_scaled , y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test , y_pred)
r2 = r2_score(y_test , y_pred)
```

```python
print("Mean_sqared_error " , mse)
print("Meaa_Absolute_Error" , mae)
print("R2_Score" , r2)
```

```python
plt.scatter(y_test , y_pred)
plt.show()
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso()
lasso_model.fit(X_train_scaled , y_train)
y_pred_lasso = lasso_model.predict(X_test_scaled)

mae = mean_absolute_error(y_test , y_pred_lasso)
mse = mean_squared_error(y_test , y_pred_lasso)
score = r2_score(y_test , y_pred_lasso)

print("Mean_abbsolute_error" , mae)
print("mean_squared_error" , mse)
print("r2_score" , score)

plt.scatter(y_test  , y_pred_lasso)
plt.show()
```

# Ridge Regression model

```python
from sklearn.linear_model import Ridge

Ridge_model = Ridge()
Ridge_model.fit(X_train_scaled , y_train)
y_pred_Ridge = Ridge_model.predict(X_test_scaled)

mae = mean_absolute_error(y_test , y_pred_Ridge)
mse = mean_squared_error(y_test , y_pred_Ridge)
score = r2_score(y_test , y_pred_Ridge)

print("Mean_abbsolute_error" , mae)
print("mean_squared_error" , mse)
print("r2_score" , score)

plt.scatter(y_test  , y_pred_Ridge)
plt.show()
```

## elasicnet Regression

```python
from sklearn.linear_model import ElasticNet

ElasticNet_model = ElasticNet()
ElasticNet_model.fit(X_train_scaled , y_train)
y_pred_ElasticNet = ElasticNet_model.predict(X_test_scaled)

mae = mean_absolute_error(y_test , y_pred_ElasticNet)
mse = mean_squared_error(y_test , y_pred_ElasticNet)
score = r2_score(y_test , y_pred_ElasticNet)

print("Mean_abbsolute_error" , mae)
print("mean_squared_error" , mse)
print("r2_score" , score)

plt.scatter(y_test  , y_pred_ElasticNet)
plt.show()
```
